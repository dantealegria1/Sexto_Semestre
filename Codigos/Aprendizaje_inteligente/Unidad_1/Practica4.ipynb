### Practica numero 2 algoritmo de perceptron simple aplicado a la flor de iris
### Cargar dos clases de la flor de iris, las caracteristicas que tomaremos en cuenta son la longitud y ancho del petalo

```python
import numpy as np
import pandas as pd
```

### Cargamos el dataset

```python
df = pd.read_csv('iris.csv',delimiter = ';',decimal=".")
print(df.shape)
print(df.head())
```

### 
Extraemos el largo sepal y el largo del petalo en las colummbas 0 y 2. Usaremos solo Setosa y versicolor

```python
x=df.iloc[0:100,[0,2]].values
y=df.iloc[0:100,4].values
#print(y)
```

```python
y=np.where(y=='setosa',-1,1) #reemplaza los valores de y, si es Iris-setosa por -1, si no por 1
print(y)
```

### Entrenamos el perceptron

```python
from sklearn.linear_model import Perceptron
Perceptron=Perceptron()
```

```python
Perceptron.fit(x, y)
```

```python
Perceptron.predict([[5.7,4.1],[5.1,1.4]])
```

### Graficamos

```python
import matplotlib.pyplot as plt
#graficas los 100 registros
plt.scatter(x[:50, 0], x[:50, 1], color='green', marker='o', label='setosa')
plt.scatter(x[50:100, 0], x[50:100, 1], color='blue', marker='x', label='versicolor')

#graficamos etiquetas
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')

#graficamos la leyenda
plt.legend(loc='upper left')
```
