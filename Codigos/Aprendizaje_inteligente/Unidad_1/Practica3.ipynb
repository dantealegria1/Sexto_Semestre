```python
#LIBRERIAS 
import numpy as np
import matplotlib.pyplot as plt
```

Datos del eje x para los siguientes graficos 

```python
x = np.linspace(0,10,100)
print(x)
```

Ejemplos

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt. figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2) 
plt.plot(x, np.cos(x))
```

En un estilo mas orientao a situaciones mas complejas

```python
fig, ax = plt.subplots(2)
#llama el petodo plot
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

Usando colores

```python
plt.plot(x, np.sin(x - 0), color='blue')        # nombre del color
plt.plot(x, np.sin(x - 1), color='g')           # código corto del color (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1
#plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Código hexadecimal (RRGGBB de 00 a FF)
#plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Tupla RGB, valores entre 0 y 1
#plt.plot(x, np.sin(x - 5), color='chartreuse'); # todos los nombres de colores HTML soportados
```

Usando colores (lineas)

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

Usando codigos

```python
plt.plot(x, x + 0, linestyle='-')
plt.plot(x, x + 1, linestyle='--')
plt.plot(x, x + 2, linestyle='-.')
plt.plot(x, x + 3, linestyle=':');
plt.plot(x, x + 3, linestyle=':');
```

Cambiando los limites de los ejes

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

Titulos

```python
plt.plot(x, np.sin(x))
plt.title("Funcion Seno(x)")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

Leyendas

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
#plt.axis('equal')
#plt.legend()
plt.legend(loc='upper right')
```

### Ejemplo flor de iris

```python
import numpy as np
import pandas as pd
df = pd.read_csv('iris.csv', delimiter=';', decimal='.')
#Numero de renglones y columnas
print(df.shape)
#Primeros 5 registros
print(df.head())
```

Elimina la variable categorica, deja las variables predictorias en x

```python
x = df.iloc[0:100, [0, 2]].values
#print(x)
```

```python
import matplotlib.pyplot as plt
#Graficar los primeros 100 registros
plt.scatter(x[:50, 0], x[:50, 1], color='red', marker='o', label='setosa')
plt.scatter(x[50:100, 0], x[50:100, 1], color='blue', marker='x', label='versicolor')

#Graficamos las etiquetas
plt.xlabel('Longitud del sepalo [cm]')
plt.ylabel('Longitud del petalo [cm]')
plt.legend(loc='upper left')

#Leyendas 
plt.show()
```

### Ejemplos Digitiso. Los vamos usar mas adelante

```python
from sklearn.datasets import load_digits
digitos = load_digits(n_class=6)
print(digitos)
```

```python
fig, ax = plt.subplots(8, 8, figsize=(6, 6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digitos.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```
